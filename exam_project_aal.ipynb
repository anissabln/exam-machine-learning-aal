```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# view the most appearing words

import re
from wordcloud import WordCloud

# Remove punctuation
songs_data['track_name_proc'] = \
songs_data['track_name'].map(lambda x: re.sub('[,.!?]', '', x))

# Convert all the words to lowercase
songs_data['track_name_proc'] = \
songs_data['track_name_proc'].map(lambda x: x.lower())


# Join the different processed bios together.
long_string = ','.join(list(songs_data['track_name_proc'].values))
# Create a WordCloud object
wordcloud = WordCloud(background_color="white", max_words=5000, contour_width=3, 
contour_color='steelblue')# Generate a word cloud
wordcloud.generate(long_string)# Visualize the word cloud
plt.figure( figsize=(20,10) )
plt.imshow(wordcloud)
plt.show();
```
